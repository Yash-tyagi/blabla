```python
import pandas as pd
import numpy as np
from skimage import io
from skimage.transform import rotate
import cv2
import os
import time
```

```python
files = os.listdir("Dataset/train/ModerateDemented/")

if('.DS_Store' in files):
    files.remove('.DS_Store')
    
    
```

```python
len(files)
```

```python
path = "Dataset/train/ModerateDemented/"
new_path = "Dataset/train/ModerateDemented_AUG/"
```

```python
def rotate_images(new_path,path,degree_of_rotation,file,count):
    img = io.imread(str(path+file))
    img = rotate(img, degree_of_rotation)
    io.imsave(new_path + 'rot'+ str(count) + '_' + str(file), img)
    
    
    
def mirror_images(new_path,path,direction,file,count):
    img=io.imread(str(path+file))
    img=cv2.flip(img,direction)
    #cv2.imwrite(new_path+'mir1_'+str(file),img)
    io.imsave(new_path + 'mir' +str(count) + '_' + str(file) , img)

    
```

```python
for file in files:
    str1 = str(file)
    str1 = str1[0:-4]
    rotate_images(new_path,path,0,file,0)
    mirror_images(new_path,path,1,file,1)
    rotate_images(new_path,path,150,file,2)
    rotate_images(new_path,path,90,file,3)
    rotate_images(new_path,path,180,file,4)
    rotate_images(new_path,path,270,file,5)
```
